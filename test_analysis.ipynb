```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create dummy retail data
num_records = 1000
data = {
    'TransactionID': np.arange(1, num_records + 1),
    'CustomerID': np.random.randint(1000, 2000, num_records),
    'ProductID': np.random.randint(100, 200, num_records),
    'Quantity': np.random.randint(1, 10, num_records),
    'Price': np.round(np.random.uniform(5.0, 100.0, num_records), 2),
    'TransactionDate': pd.date_range(start='2023-01-01', periods=num_records, freq='H'),
    'Discount': np.round(np.random.uniform(0.0, 20.0, num_records), 2)

}

df = pd.DataFrame(data)
df.head(2)
```

```python
# total sales
df['TotalSales'] = df['Quantity'] * df['Price'] - df['Discount']
df.head(2)
```

```python
# plot discount vs total sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='TotalSales', data=df)
plt.title('Discount vs Total Sales')
plt.xlabel('Discount ($)')
plt.ylabel('Total Sales ($)')
plt.show()
```
