```python
import pandas as pd
import numpy as np
```

```python
# lets generate the population data
population = np.random.randint(0, 100, size=1000)
population_mean = np.mean(population)
print(f"Population Mean: {population_mean}")
```

```python
# lets create the sample data
sample_size = 50
sample_1 = np.random.choice(population, size=sample_size, replace=False)
sample_2 = np.random.choice(population, size=sample_size, replace=False)
sample_3 = np.random.choice(population, size=sample_size, replace=False)    
sample_4 = np.random.choice(population, size=sample_size, replace=False)
sample_5 = np.random.choice(population, size=sample_size, replace=False)

# calculate the sample means
sample_means = [np.mean(sample_1), np.mean(sample_2), np.mean(sample_3), np.mean(sample_4), np.mean(sample_5)]
print(f"Sample Means: {sample_means}")
```

```python
# lets create variable age with random values between 0 to 100 for 1000 entries
age = np.random.randint(0, 100, size=1000)
# create a dataframe with age and population
df = pd.DataFrame({'age': age, 'population': population})
df.head(2)
```

```python
# lets use
hist, bins = np.histogram(df['age'], bins=10)
print(f"Histogram: {hist}")
print(f"Bins: {bins}")
# plot the histogram
import matplotlib.pyplot as plt
plt.hist(df['age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
# lets go for KDE plot
import seaborn as sns
sns.kdeplot(df['age'], shade=True)
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Age Distribution - KDE Plot')
plt.show()
```
