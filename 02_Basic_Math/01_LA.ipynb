```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

x = np.array([0, 1, 2, 3, 4, 5])
y1 = -4 * x + 3
y2 = 3 * x + 3
plt.plot(x, y1, label='y = -4x + 3')
plt.plot(x, y2, label='y = 3x + 3')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lines: y = 2x + 3 and y = 3x + 3')
plt.grid()
plt.legend()
plt.show()
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=100, centers=2, random_state=6)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2') 
plt.title('Scatter Plot of Two Classes')

# Add a line with slope
slope = -1.21
intercept = 2
x_vals = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
point = np.array([10.2, 1.3])  # Example data point
plt.scatter(point[0], point[1], color='green', s=100, label='Test Point')

# Calculate the line value at point's x-coordinate
line_value = slope * point[0] + intercept
plt.scatter(point[0], line_value, color='red', s=100, label='Line Value')
# Draw a vertical line from the point to the line
plt.plot([point[0], point[0]], [point[1], line_value], 'r--', label='Distance to Line')
y_vals = slope * x_vals + intercept
plt.plot(x_vals, y_vals, 'k--', label=f'y = {slope}x + {intercept}')

plt.grid()
plt.legend()
plt.show()
```
