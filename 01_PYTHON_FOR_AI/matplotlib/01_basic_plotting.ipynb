```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# lets create some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Basic Plotting
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# lets create some categorical data
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

# Basic Bar Plot
plt.bar(categories, values)
plt.title("Basic Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# lets plot pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [10, 15, 7, 12]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Basic Pie Chart")
plt.show()
```

```python
# lets create sales data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [200, 220, 250, 270, 300, 320, 350, 370, 400, 420, 450, 480]
}
df = pd.DataFrame(data)
df.head(10)
```

```python
# lets plot it 
plt.plot(df['Month'], df['Sales'], marker='o')
plt.title("Monthly Sales Data")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# lets create complex data : banking data
np.random.seed(0)
n = 1000
data = {
    'Age': np.random.randint(18, 70, n),
    'Income': np.random.randint(20000, 100000, n),
    'CreditScore': np.random.randint(300, 850, n),
    'LoanAmount': np.random.randint(1000, 50000, n),
    'Defaulted': np.random.choice([0, 1], n, p=[0.8, 0.2]),
    # create categorical variable
    'Gender': np.random.choice(['Male', 'Female'], n, p=[0.5, 0.5]),
    'City': np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'], n)
}
df_bank = pd.DataFrame(data)
df_bank.head()
```

```python
# income defaulted
df_bank.groupby('Defaulted')['Income'].mean()
plt.figure(figsize=(10, 6))
plt.bar(['Not Defaulted', 'Defaulted'], df_bank.groupby('Defaulted')['Income'].mean(), color=['green', 'red'])
plt.title('Average Income by Default Status')
plt.xlabel('Default Status')
plt.ylabel('Average Income')
plt.grid(axis='y')
plt.show()
```

```python
# for the same income I can plot distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df_bank, x='Income', hue='Defaulted', bins=30, kde=True, palette={0: 'green', 1: 'red'}, alpha=0.6)
plt.title('Income Distribution by Default Status')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
# lets talk about credit score
plt.figure(figsize=(10, 6))
plt.scatter(df_bank['CreditScore'], df_bank['LoanAmount'], c=df_bank['Defaulted'], cmap='bwr', alpha=0.6)
plt.title('Credit Score vs Loan Amount')
plt.xlabel('Credit Score')
plt.ylabel('Loan Amount')
plt.colorbar(label='Defaulted (0=No, 1=Yes)')
plt.grid()
plt.show()
```

```python
# lets create time series
# Generate time series data
date_range = pd.date_range(start='2020-01-01', periods=100, freq='D')
time_series_data = pd.DataFrame({
    'Date': date_range,
    'Value': np.random.randn(100).cumsum()  # Cumulative sum of random values
})
time_series_data.head()

# lets plot it
plt.figure(figsize=(10, 6))
plt.plot(time_series_data['Date'], time_series_data['Value'], marker='o', linestyle='-', color='b')
plt.title("Time Series Data")
plt.xlabel("Date")
plt.ylabel("Value")
plt.grid()
plt.xticks(rotation=45)
plt.show()
```
